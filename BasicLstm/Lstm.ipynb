```python
# Importing Statements
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
Data = [[[i+j] for j in range(5)] for i in range(100)]
target = [[i+5] for i in range(100)]
```

```python
Data = np.array(Data, dtype=float)
target = np.array(target, dtype=float)
```

```python
target.shape, Data.shape
```

```python
Data /= 100
target /= 100
```

```python
X_train, X_test, y_train, y_test = train_test_split(Data, target, random_state=4, test_size=0.2)
```

```python
model = Sequential([
    LSTM((1), batch_input_shape=(None, 5, 1), return_sequences=False)
])
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=400, batch_size=32, validation_data=(X_test, y_test))
```

```python
result = model.predict(X_test)
```

```python
plt.scatter(range(20), result)
plt.scatter(range(20), y_test)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.show()
```
